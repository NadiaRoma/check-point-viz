```python
def plot_correlation_map( df ):

    corr = df.corr()

    s , ax = plt.subplots( figsize =( 12 , 10 ) )

    cmap = sns.diverging_palette( 220 , 10 , as_cmap = True )

    s = sns.heatmap(corr, cmap = cmap, square=True, cbar_kws={ 'shrink' : .9 }, ax=ax, annot = True, annot_kws = { 'fontsize' : 12 })
    
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
plot_correlation_map(df)
```

```python
#we have created a function that allows to visualise the correlation between the different features. The closer the correlation
#is towards 1, the more the variables are correlated
#We started by defining the variable Corr.the variables ax and s are X and Y. The Subplot function allows you to visualize a 
#graph that presents the dataframe in the form of a table. The variable Cmap allows to visualize the rectangle on the right 
#with the color code.
#The correlation varies between -1 and 1 (from blue to red). The more the correlation tends towards 1 the more the two variables
#are proportionally linked. The closer the correlation is to -1 the more the variables are inversely related.
#The analysis of the graph allows us to see that Pclass and Fare are inversely correlated. SbSp and Pclass are decorrelated.
#Parch and SbSp are correlated
```
